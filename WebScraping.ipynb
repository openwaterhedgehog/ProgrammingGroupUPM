```python
# import necessary modules for web scraping using Selenium and csv moduler for handling CSV files
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
```

```python
# creates an instance of the Chrome webdriver that can be used to automate interactions with a Chrome browser
wd = webdriver.Chrome()
```

```python
# instructs the webdriver to the URL specified as the argument
wd.get('https://www.investing.com/funds/amundi-msci-wrld-ae-c-historical-data')
```

```python
# maximize window to ensure possible interaction with all elements
# scroll down to where table is positioned in the browser
wd.maximize_window()
wd.execute_script("window.scrollTo(0,300)")
```

```python
# XPATH expressions for locating elements
date_xpath =  "/html/body/div[5]/section/div[8]/div[3]/div/div[2]/span"
start_xpath =  "//*[@id=\"startDate\"]"
end_xpath = "//*[@id=\"endDate\"]"
startDate = "01/01/2020"
endDate = "12/31/2020" 
apply_xpath =  "//*[@id=\"applyBtn\"]"
table_xpath = "//*[@id=\"results_box\"]"
```

```python
# uses the Selenium library to select the date range specified to scrape data from the table

date_picker = WebDriverWait(wd,30).until(EC.element_to_be_clickable((By.XPATH,date_xpath)))

wd.execute_script("arguments[0].click();", date_picker)
    
start_date = wd.find_element(By.XPATH,start_xpath)
start_date.clear()
start_date.send_keys(startDate)
end_date = wd.find_element(By.XPATH,end_xpath)
end_date.clear()
end_date.send_keys(endDate)
apply_button = wd.find_element(By.XPATH,apply_xpath)
wd.execute_script("arguments[0].click();", apply_button)

table_data = WebDriverWait(wd, 30).until(EC.presence_of_element_located((By.XPATH, table_xpath)))
table_data_header = WebDriverWait(wd, 30).until(EC.presence_of_element_located((By.XPATH, "//*[@id=\"curr_table\"]/thead/tr")))
table_data_body = WebDriverWait(wd, 30).until(EC.presence_of_element_located((By.XPATH, "//*[@id=\"curr_table\"]/tbody")))
```

```python
print(table_data_header.text)
print(table_data_body.text)
print(type(table_data_header.text))
print(type(table_data_body.text))
```

```python
# uses the csv module to write data from a web page table to a CSV file.
table_data_header_list = str.split(table_data_header.text)
table_data_header_list[-2:] = [' '.join(table_data_header_list[-2:])]
rows = table_data_body.text.strip().split('\n')
data = [row.split() for row in rows]
data_combined = [[sublist[0] + ' ' + sublist[1] + ' ' + sublist[2]] + sublist[3:] for sublist in data]

with open('test.csv', 'w') as f:
    write = csv.writer(f)
    write.writerow(table_data_header_list)
    write.writerows(data_combined)


```
